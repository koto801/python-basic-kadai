```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")

# インスタンスを複数生成してリストに追加
humans = [
    Human("田中", 25),
    Human("佐藤", 17),
    Human("鈴木", 20),
    Human("高橋", 15),
]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
